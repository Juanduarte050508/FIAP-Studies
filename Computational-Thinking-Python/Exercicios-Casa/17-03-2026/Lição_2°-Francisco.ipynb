```python
#Classificação de crédito : solicite renda mensal e score de crédito. Crie as seguintes regras:

#aprovado - renda => 3000 and score >= 700
#Analise manual - renda => 2000 and score >=600
#caso contrario- negado 

Renda_mensal= float (input("Digite sua renda mensal"))
score= float (input("Digite seu score"))

if Renda_mensal >= 3000 and score >= 700:
    print("Aprovado!")

elif Renda_mensal >= 2000 and score >= 600:
    print("Analise manual")

else: print("Negado!")

```
