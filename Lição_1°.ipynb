```python
#lição 1, Àrea do Circulo//

#Variaveis
pi=3.14
raio = float(input("Insira o raio do circulo"))
R2= raio*raio

#Cálculo da área
Área = pi*R2

#Esrutura de saída
print ("A Área do circulo é", Área)
```

```python
#lição 2, área do retangulo//

#Variaveis
lado1= float(input('Digite o lado 1 do retangulo:'))
lado2= float(input("Digite o lado 2 do rentangulo:"))

#Calculo do retangulo
Area = lado1*lado2

#Estrutura de saída 
print("Área do retangulo", Area)
```

```python
#lição 3, área do triangulo//

#Variaveis
Base= float(input('Digite a base do triangulo'))
Altura= float(input("Digite a altura"))

#Calculo do triangulo
Area = (Base*Altura)/2

#Estrutura de saída
print('Area do triangulo', Area)
```
