```python


# Inserir variavel de entrada = input = caixa de texto 

# //Variáveis de entrada//
# float = valores decimais 1,5 por exemplo
# int = numeros inteiros
# Str(string) = Qualquer coisa que não esteja relacionada com numero, como palavra e texto

#//Variaveis de saida//
# Print = mostrar a saida, por meio do texto 
n1 = float(input("Digite a priemira nota:"))
n2 = float(input("Digite a segunda nota:"))
n3 = float(input("Digite a terceira nota:"))

#Cálculo da média
nota_final= n1+n2+n3
print (nota_final)

#//Estrutura de sáida//
print ("Primeira nota", n1)
print ("Segunda nota", n2) 
print ("Terceira nota", n3)
print ("Media", nota_final)

#//Estrutura de condição//
if nota_final >= 6:
    print("Aprovado!")
else:
    print("Reprovado!")
```
